MATPLOTLIB

IMPORTAMOS LIBRERIAS A USAR

```python
import numpy as np 
from numpy import random
import matplotlib.pyplot as plt
```

LINEPLOTS

```python
x = [2000,2001,2002,2003,2004,2005]
poblacionA = random.randint(50,70,6)
poblacionB = random.randint(50,70,6)
poblacionC = random.randint(50,70,6)

#Tama√±o de la venta
plt.figure(figsize=(10,6))

plt.plot(x,poblacionA,marker="o",label="China",color="g")
plt.plot(x,poblacionB,marker="o",label="Argentina",color="y")
plt.plot(x,poblacionC,marker="o",label="Colombia",color="r")


#Titulos de los ejes x y y
plt.ylabel("Poblacion")
plt.title("Evaluador de poblacion")

# Leyenda y ubicacion
plt.legend(loc=4)

#Mostrar cuadricula
plt.grid()

#Mostrando grafico
plt.show()
```

BARPLOTS

```python
#Ponemos las categorias del eje x y y
x = ["Frijol","Cafe","Arveja","Mora","Papa"]
y = random.randint(0,15,5)

#Titulo y titulos del eje x y y 
plt.title("Cultivos en santa leticia")
plt.ylabel("N¬∞ de personas que lo siembran")

#Cargamos el diagrama de barra y le pasamos los argumentos
plt.bar(x,y,color="r",edgecolor="black")
#Mowtramos el diagrama

plt.grid()
plt.show()
```

```python
#Ponemos las categorias del eje x y y
x = np.arange(4)
y = random.randint(0,10,4)
y2 = random.randint(0,10,)

#Titulo y titulos del eje x y y 
plt.title("Deportes")

#Cargamos el diagrama de barra y le pasamos los argumentos
plt.bar(x + 0.05,y, width=0.25 , label="Clase 1")
plt.bar(x + 0.30,y2,width=0.25 , label="Clase 2")

plt.legend(loc=1)

#Estableciendo nombres del eje x
plt.xticks(x + 0.30, ["F√∫tbol","B√°squetbol","V√≥leibol","Tenis"])

#Mowtramos el diagrama
plt.grid(color='k', linestyle='-', linewidth=0.2)
plt.show()
```

PIECHARTS

```python
x = ["F√∫tbol","B√°squetbol","V√≥leibol","Tenis"]
y = random.randint(1,100,4)

#Colores en hexadecimal para el pie
colores = ["#EE6055","#60D394","#AAF683","#FFD97D","#FF9B85"]

# Para separar cada cuadro del otro
desfase = (0.01, 0.01, 0.01, 0.01)

plt.pie(y,labels=x,colors=colores,explode=desfase,autopct="%0.1f %%")

plt.show()
```

BOXPLOTS

```python
edad = random.randint(1,125,100)
plt.boxplot(edad)

plt.grid()
plt.show()
```

SCATTERPLOT

```python
a = random.randint(1,100,10)
b = random.randint(1,100,10)

plt.scatter(a,b,color="r")
plt.grid()
plt.show()
```

SUBPLOT

```python
x = [2000,2001,2002,2003,2004,2005]
poblacionA = random.randint(50,70,6)
poblacionC = random.randint(50,70,6)

#Pasamos de argumentos filasxcolumnas y que compartan el mismo eje y
fig, ax = plt.subplots(1,2,sharey=True)

ax[0].plot(x,poblacionA,color="r",label="China")
ax[0].legend(loc=2)
ax[0].grid()

ax[1].plot(x,poblacionC,color="b",label="China")
ax[1].legend(loc=2)
ax[1].grid()

plt.show()
```
