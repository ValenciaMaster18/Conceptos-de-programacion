#**JavaScript**

JavaScript es un lenguaje de scripting multiplataforma orientado a objetos que se utiliza para hacer que las páginas web sean interactivas (por ejemplo, con animaciones complejas, botones en los que se puede hacer clic, menús emergentes, etc.).

#**Gramatica y tipos**

#**Basico**

> JavaScript toma prestada la mayor parte de su sintaxis de Java, C y C++, pero también ha sido influenciado por Awk, Perl y Python.

> JavaScript distingue entre mayúsculas y minúsculas.

> En JavaScript, las instrucciones se denominan instrucciones y están separadas por punto y coma (;).

#**Comentarios**

*   Un identificador de JavaScript suele comenzar con una letra, un guión bajo () o un signo de dólar ().

#**Alcance de las variables**

> Cuando se declara una variable fuera de cualquier función, se denomina variable global, ya que está disponible para cualquier otro código del documento actual. 

> Cuando se declara una variable dentro de una función, se denomina variable local, porque sólo está disponible dentro de esa función.

#**Conversion de tipos de datos**

JavaScript es un lenguaje de tipado dinámico. Esto significa que no es necesario especificar el tipo de datos de una variable al declararla. También significa que los tipos de datos se convierten automáticamente según sea necesario durante la ejecución del script.

#**Declaraciones de interrupción**

La instrucción break asociada a cada case garantiza que el programa se interrumpe una vez ejecutada la instrucción coincidente y, a continuación, continúa la ejecución en la instrucción siguiente. 

**Si se omite, el programa continúa la ejecución dentro de la instrucción (y evaluará la siguiente, y así sucesivamente).**
